```python
#Import packages
import pandas as pd
import os as os
import numpy as np
import seaborn as sns
%matplotlib inline
import matplotlib.pyplot as plt
import copy
from scipy.stats import iqr
from datetime import datetime, timedelta, date

pd.set_option('display.max_rows', 500)
```

```python
df = pd.read_csv(r"C:\Users\Tom-T\Google Drive\Queen's MMA\Analytics for Financial Markets - MMA 823\Project\2019_2020_Sept_Tweets.csv", encoding = "ISO-8859-1")

df.head()
```

```python
#Remove the 'Replying to' at the start of tweets
df['Comments'] = df['Comments'].str.replace('Replying to','')
```

```python
#Remove returns
df['Comments'] = df['Comments'].str.replace('\n',' ')

#Remove Twitter handles in responses
df['Comments'] = df['Comments'].str.replace('@[A-Za-z0-9]+', '',regex=True)

#Remove Special Characters and Emojis
df['Comments'] = df['Comments'].str.replace('[^ a-zA-Z0-9]', '',regex=True)
```

```python
df.head()
```

```python
#Fill NaN values for Replies/Retweets/Likes with 0
df = df.fillna(0)
```

```python
#Convert timestamp column to datetime datatype
df['Timestamp'] = df['Timestamp'].str.slice(stop=10)

df['Timestamp'] = pd.to_datetime(df['Timestamp'],format='%Y-%m-%d')
```
