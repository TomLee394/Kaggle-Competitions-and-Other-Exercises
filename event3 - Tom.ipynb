# Build Pipeline

```python
from sklearn.model_selection import train_test_split

X = df['sms_text']
y = df['spam']

X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.1,stratify=y)
```

```python
type(X_train)
X_train.shape
X_train.head()

type(y_train)
y_train.shape
y_train.head()
```

```python
# Simple preprocessor.
# Input is a single document, as a single string.
# Otuput should be a single document, as a single string.
def my_preprocess(doc):
    
    # Lowercase
    doc = doc.lower()
    
    # Remove URLs
    doc = re.sub(r'http\S+', 'http ', doc)
        
    #Creation of a call to action type of feature
    
    if re.search(r'.*call \d+.*', doc) is not None:
          doc = doc + ' Call_To_Action_Ind'

    if re.search(r'.*call.*\d+.*', doc) is not None:
          doc = doc + ' Call_To_Action_Ind'
            
    if re.search(r'.*Call.*\d+.*', doc) is not None:
          doc = doc + ' Call_To_Action_Ind'
            
    if re.search(r'.*Call.*\d+.*', doc) is not None:
          doc = doc + ' Call_To_Action_Ind'       
            
    if re.search(r'.*text \d+.*', doc) is not None:
          doc = doc + ' Call_To_Action_Ind'  
            
    if re.search(r'.*text.* \d+.*', doc) is not None:
          doc = doc + ' Call_To_Action_Ind'  
            
    if re.search(r'.*sex.*', doc) is not None:
          doc = doc + ' Call_To_Action_Ind'  
            
    if re.search(r'.*NOW.*', doc) is not None:
          doc = doc + ' Call_To_Action_Ind'  
            
    if re.search(r'.*URGENT.*', doc) is not None:
          doc = doc + ' Call_To_Action_Ind'  
    
    if re.search(r'.*FREE.*', doc) is not None:
          doc = doc + ' Call_To_Action_Ind' 
    # TODO: What else?

    return doc
```

```python
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import MultinomialNB

vectorizer = CountVectorizer(preprocessor = my_preprocess)
dt = MultinomialNB()


pipe = Pipeline([('cv', vectorizer), ('clf', dt)])


# Other things to try above:
# - More/Different preprocessing steps
# - Different hyperparameter values for CountVectorizer
# - Different hyperparameter values for DecisionTreeClassifier
# - Different ML algorithms
# - Additional feature engineering (can be added to Pipeline via sklearn.pipeline.FeatureUnion)
```

```python
pipe.fit(X_train, y_train)
```

# Estimate Model Performance

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report, f1_score

pred_val = pipe.predict(X_val)

print("Confusion matrix:")
print(confusion_matrix(y_val, pred_val))

print("\nF1 Score = {:.5f}".format(f1_score(y_val, pred_val, average="micro")))

print("\nClassification Report:")
print(classification_report(y_val, pred_val))
```
